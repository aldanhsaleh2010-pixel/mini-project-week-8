# Inputs Definition 📍

```python
import numpy as np
import pandas as pd

X = np.array([
    [1.0, 0.5],
    [2.0, -1.0],
    [0.0, 3.0],])


W = np.array([
    [0.2, -0.1, 0.5],
    [0.7, 0.3, -0.2],])


b = np.array([0.1, 0.0, -0.3])
```

# Loading Data from CSV files 🔧


```python
X_linear = pd.read_csv("/kaggle/input/datasets/dana3333/40linear/day40_linear.csv")
W_weights= pd.read_csv("/kaggle/input/datasets/dana3333/weights40/day40_weights.csv")
b_bias = pd.read_csv("/kaggle/input/datasets/dana3333/bias40/day40_bias.csv")
```

# step 1 , Build a simple linear transformation🧐

#### In this step, we understand the linear transformation equation and verify that the matrix dimensions are compatible before computing Y.🔍

```python
print("\n x shape \n",X.shape)
print("\n w shape \n",W.shape)
print("\n b shape \n",b.shape)
```

# Compute the linear transformation📔
#### We compute this step because linear transformation is the foundation of many machine learning models📍

```python
Y = X @ W + b

print("Y shape:", Y.shape)
print("\nY:\n", Y)
```

## We used vectorized computation (X @ W + b) instead of loops because it is faster and more efficient in NumPy.❗

# Broadcasting with Norm-Based Scalin ➖➗🟰
#### We scale the data because some features may have very large values while others have very small values, which can negatively affect model performance.📉

```python
Y_mean = Y.mean(axis=0, keepdims=True)
Y_std = Y.std(axis=0, keepdims=True)

Y_norm = (Y - Y_mean) / (Y_std + 1e-8)

print("Y_norm mean:", Y_norm.mean(axis=0))
print("Y_norm std:", Y_norm.std(axis=0))

print(Y_norm.mean(axis=0))
print(Y_norm.std(axis=0))
```

# Matrix Operations for Feature Engineering✖️


```python
X_raw = np.array([
    [1., 2., 3.],
    [0.5, 1.5, 2.],
    [2., 0.5, 1.]
])

A = np.array([
    [1., 0.],
    [0.5, 0.5],
    [0., 1.]
])

X_new = X_raw @ A

print("X_raw shape:", X_raw.shape)
print("X_new shape:", X_new.shape)
```

#### We use matrix multiplication to create new features as linear combinations of the original features, which helps in dimensionality reduction and feature engineering.⭕❗

# Linear Transformation Pipeline⏮️

```python
X_mean = X.mean(axis=0, keepdims=True)
X_std = X.std(axis=0, keepdims=True)
X_stdzd = (X - X_mean) / (X_std + 1e-8)

Y = X_stdzd @ W + b

Y_unit = Y / (np.linalg.norm(Y, axis=1, keepdims=True) + 1e-8)

print("X_stdzd shape:", X_stdzd.shape)
print("Y shape:", Y.shape)
print("Y_unit shape:", Y_unit.shape)
```

#### We built a full linear transformation pipeline using vectorized operations without loops.💥
